```python
urls = ['https://inshorts.com/en/read/sports',
        'https://inshorts.com/en/read/politics',
        'https://inshorts.com/en/read/technology']
def build_dataset(urls):
  news_data = []
  for url in urls:
    news_category = url.split('/')[-1]
    #print(news_category)
    data = requests.get(url)
    soup = BeautifulSoup(data.content)
    #print(soup)
  
    news_articles = [{'news_headline':headline.find('span',attrs={"itemprop":"headline"}).string,
                      'news_article':article.find('div',attrs={"itemprop":"articleBody"}).string,
                      'news_category':news_category}

                    for headline,article in zip(soup.find_all('div',class_=["news-card-title news-right-box"]),  soup.find_all('div',class_=["news-card-content news-right-box"]))]
    news_articles = news_articles[0:20]
    news_data.extend(news_articles)
     
  df= pd.DataFrame(news_data)
  df = df[['news_headline','news_article','news_category']]
  return df
```

```python
df = build_dataset(urls)
df
```

**PREPROCESSING**

Cleaning the textual data

```python
import nltk
nltk.download('stopwords')
stopword_list = nltk.corpus.stopwords.words('english')
print(stopword_list)
print(len(stopword_list))
```

```python
#removing two important stop words (no and not)

stopword_list.remove('no')
stopword_list.remove('not')
len(stopword_list)
```

```python
#remove html tags
def html_tag(text):
   soup = BeautifulSoup(text,"html.parser")
   new_text = soup.get_text()
   return new_text

html_tag('<html><h2> Some important information </h2></html>')
```

```python
#expand contractions

!pip install contractions
import contractions
def con(text):
  expand = contractions.fix(text)
  return expand

con("y'all can't expand I'd think")
```

```python
#removing special characters

import re
def remove_sp(text):
  pattern = r'[^A-Za-z0-9\s]'
  text = re.sub(pattern,'',text)
  return text

print(remove_sp("Well this was fun! What do you think? 123#"))
print(remove_sp("@#$abc"))
```

```python
from nltk.tokenize.toktok import ToktokTokenizer
tokenizer = ToktokTokenizer()
```

```python
#removing stop words
# Tokenization -  splitting a sentence into list

def remove_stopwords(text):
  tokens = tokenizer.tokenize(text)    #considering tokens
  tokens = [token.strip() for token in tokens]
  filtered_tokens = [token for token in tokens if token not in stopword_list]       # checking whether it is in stopwords_list or not if there removing from stopwords_list
  filtered_text = ' '.join(filtered_tokens)
  return filtered_text
```

```python
remove_stopwords(" The, and, if are all stopwords and even not")
```

```python
# remove punctuation
import string
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
  
input_str = "Hey, did you know that the summer break is coming? Amazing right !! It's only 5 more days !!"
remove_punctuation(input_str)
```

```python
# Remove numbers
def remove_numbers(text):
    result = re.sub(r'\d+', '', text)
    return result
  
input_str = "There are 3 balls in this bag, and 12 in the other one."
remove_numbers(input_str)
```

```python
# remove whitespace from text
def remove_whitespace(text):
    return  " ".join(text.split())
  
input_str = "   we don't need   the given questions"
remove_whitespace(input_str)
```

```python
# REMOVING
#1. Converting to Lower Case
#2. HTML tags
#3. Contractions
#4. Special characters
#5. stop words
#6. Numbers
#7. white space
```

```python
df.news_headline = df.news_headline.apply(lambda x:x.lower())
df.news_article = df.news_article.apply(lambda x:x.lower())


df.news_headline = df.news_headline.apply(html_tag)
df.news_article = df.news_article.apply(html_tag)

df.news_headline = df.news_headline.apply(con)
df.news_article = df.news_article.apply(con)

df.news_headline = df.news_headline.apply(remove_sp)
df.news_article = df.news_article.apply(remove_sp)

df.news_headline = df.news_headline.apply(remove_stopwords)
df.news_article = df.news_article.apply(remove_stopwords)

df.news_headline = df.news_headline.apply(remove_numbers)
df.news_article = df.news_article.apply(remove_numbers)

df.news_headline = df.news_headline.apply(remove_whitespace)
df.news_article = df.news_article.apply(remove_whitespace)

df

```

```python
score_neu=[]
score_pos = []
score_neg = []
for i in range(0,df.shape[0]):
  score = vs.polarity_scores(df.iloc[i][1])
  score1 = score['neu']
  score_neu.append(score1)
  score2 = score['pos']
  score_pos.append(score2)
  score3 = score['neg']
  score_neg.append(score3)
```
